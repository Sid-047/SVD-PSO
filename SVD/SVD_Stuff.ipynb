```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
A = imread("N:\\SVDnPSO_Work\\pic.png")
X = np.mean(A, axis = -1)
X
```

```python
plt.imshow(X, cmap = "gray")
```

```python
U, S, VT = np.linalg.svd(X, full_matrices = False)
S = np.diag(S)
```

```python
plt.figure(0)
plt.semilogy(np.diag(S))
plt.title("Singular Values")
plt.show()

plt.figure(1)
plt.plot(np.cumsum(np.diag(S)/np.sum(np.diag(S))))
plt.title("Singular Values: Cummulative Sum")
plt.show()
```

```python
for r in [1,5,10,20,50,100,200,250]:
    print(">>>>>>>>", r)
    X_approx = U[:, :r] @ S[:r, :r] @ VT[:r, :]
    img = X_approx - 256
    plt.imshow(img, cmap = "gray")
    plt.title("Rank {}".format(r))
    plt.show()
    img_ = Image.fromarray((img * 255).astype(np.uint8))
    img_.save("N:\\SVDnPSO_Work\\Rank_pics\\{}.png".format(r))
```
